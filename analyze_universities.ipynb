```python
import pandas as pd
from bs4 import BeautifulSoup as bs
```

## Import data

```python
uni = pd.read_excel('universities.xlsx')
uni.head()
```

## Clean data

```python
uni['price_usd'] = uni['price_usd'].apply(lambda x: x if isinstance(x, int) else 0)
uni['location'] = uni['location'].fillna('').str.lower()
uni['country'] = uni['location'].apply(lambda x: x.split(',')[1].strip() if x != '' else x)
uni['languages'] = uni['languages'].fillna('').apply(
    lambda x: x.replace('[', '').replace(']', ''))
uni['pace'] = uni['pace'].fillna('').apply(
    lambda x: x.replace('[', '').replace(']', ''))

uni.head()
```

```python
countries_to_keep = ['united kingdom', 'france', 'italy', 'germany', 'spain', 'portugal', 'netherlands', 'belgium']

uni = uni[uni.country.isin(countries_to_keep)]

uni.head()
```

```python
uni = uni[uni.pace == "'Full-time'"]
```

```python
less_7000 = uni.query('price_usd <= 7000')
```
